Podemos darle el prompt que queramos, incluso haciendolo a través de un loop, para obtener de forma autómatica respuestas a varias preguntas:

```python
reviews = pd.read_excel('../data/reviews.xlsx')
reviews.info()
```

```python
prompt_list = reviews['text'].tolist()
```

```python
pred = []
for i in prompt_list:
    completion = openai.chat.completions.create(model="gpt-3.5-turbo",
                    messages=[
                        {"role": "system", "content": "You are an Amazon employee in charge of labeling the sentiment of the reviews left in your products. You are tasked to only anser 'Positive' when the review is positive, 'Negative' if it is a negative comment or has a negative connotation and 'Neutral' if its both, or difficult to determine."},
                        {"role": "user", "content": i}
                        ]
                )
    rev = completion.choices[0].message.content.strip()
    print(rev)

    pred.append(rev)
    time.sleep(1)
    

pred
```

```python
reviews['pred_gpt'] = pred
```
