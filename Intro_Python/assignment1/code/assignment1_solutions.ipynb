#### 1) Setting up the list of dates:

1.1) Import the file 'dates_experiment.xlsx' located in 'data/input/' folder. Use the function .head() to take a look at it:

```python
df_dates = pd.read_excel('../data/input/dates_experiment.xlsx')
df_dates.head()
```

1.2) Notice that the dates in this file go from 20/06/2022 to 14/12/2022 including weekends. The reports were only sent from 2022-10-16 to 2022-10-27 only from Sunday to Thursday. 

- Filter the data frame to obtain one with the dates that the report was sent.
- Create a list with the dates of the report.

Hint: Use the weekday() function to filter the Data Frame by day of the week.

```python
df_dates = df_dates[(df_dates['date'] > '2022-10-15') & (df_dates['date'] < '2022-10-28')]
df_dates
```

```python
# Filtering Weekends using weekday(): 
# Google: 'How to obtain the weekday from a date pandas' and you have three options
# https://stackoverflow.com/questions/28009370/get-weekday-day-of-week-for-datetime-column-of-dataframe

df_dates['weekday'] = df_dates['date'].apply(lambda x: x.weekday())
df_dates = df_dates[(df_dates['weekday'] != 4) & (df_dates['weekday'] != 5)].reset_index(drop=True)
df_dates.drop(['weekday'], axis=1, inplace=True)
df_dates
```

```python
# Export list as string of dates:
list_dates = list(df_dates['date'].astype(str))
list_dates
```

1.3) EXTRA POINT: Build a function that:

- Takes as input a data frame, the name of the column with the dates, a start and a final date.
- Filters the data frame by those two dates.
- Returns a list with the filtered dates.

```python
def get_time_list(df, var_date = 'date', initial = '2022-10-15', end = '2022-10-28'):
    df = df[(df[var_date] > initial) & (df[var_date] < end)]
    list_dates = list(df[var_date].astype(str))
    
    return list_dates
```

```python
# Using the function:
df_dates_prueba = pd.read_excel('../data/input/dates_experiment.xlsx')
list_prueba = get_time_list(df_dates_prueba, initial = '2022-10-18', end = '2022-10-28')
list_prueba
```

#### 2) Aggregating Facebook

2.1) Notice that in each folder there is a file named 'facebook.xlsx'. Loop through these folders using the list you just created, import the 'facebook.xlsx' files and append them one in top of the other to generate an aggregated data set with all the posts from this report sent over this period of time:

```python
# Using append you get all this warnings since that function is going to be removed in the future, but it works
df_final = pd.DataFrame()
for date in list_dates:
    df = pd.read_excel(f'../data/{date}/facebook.xlsx')
    df_final = df_final.append(df)
    
df_final = df_final.reset_index(drop=True)
df_final.info()
```

```python
def create_folder(directory):
    """Create folder"""
    # Create target Directory if don't exist
    if not os.path.exists(directory):
        os.mkdir(directory)
        print("Directory ", directory, " created ")
    else:
        print("Directory ", directory, " already exists")
        
create_folder('../data/facebook_out/')

df_final.to_parquet('../data/facebook_out/agg_facebook.parquet.gzip')
```

#### 3) Facebook Top Posts

3.1) Notice that in each folder there are 6 files named 'facebook_top_posts_{country}_verifiable.xlsx' and 'facebook_top_posts_{country}_not_verifiable.xlsx', where country are KE (Kenya), ZA (South Africa) and NG (Nigeria)

- Loop through these folders using the dates list, the type list and the country list provided already to you.
- Import the files and add a column named 'verfiability' which will be a dummy that is 1 if the report is 'verifiable' and 0 if 'not_verifiable', and a column with the country from which this report is from.
- Append them one in top of the other to generate an aggregated data set with all the posts from this report sent over this period of time:

Hint: Do a nested loop across *the three* lists, with 3 nested fors

```python
type_list = ['verifiable', 'not_verifiable']
country_list = ['KE', 'ZA', 'NG']
```

```python
df_stage = pd.DataFrame()
for status in type_list:
    df_status = pd.DataFrame()
    for country in country_list:
        df_country = pd.DataFrame()
        for fecha in list_dates:
            df = pd.read_excel(f'../data/{fecha}/facebook_top_posts_{country}_{status}.xlsx')
            df['country'] = country
            if status == 'verifiable':
                df['verifiability'] = 1
            else: 
                df['verifiability'] = 0
            df_country = df_country.append(df)
        df_status = df_status.append(df_country)
    df_stage = df_stage.append(df_status)
    
df_stage = df_stage.reset_index(drop=True)
df_stage
```
