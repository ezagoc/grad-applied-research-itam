```python
from geopy.geocoders import Nominatim
import pandas as pd
import numpy as np
import time
from tqdm import tqdm
from time import sleep
from random import randint
import logging
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut, GeocoderServiceError

# Defining the function:
user_agent = 'user_me_{}'.format(randint(10000,99999))
geolocator = Nominatim(user_agent=user_agent)

# This is a more elaborate function I created to stop the API of blocking me or giving me an error

def reverse_geocode(geolocator, loc, sleep_sec):
    try:
        location = geolocator.geocode(loc)
        if location == None:
            df =  pd.DataFrame([[loc, None, None]], 
                               columns=['location', 'latitude', 'longitude'])
        else:
            df =  pd.DataFrame([[loc, location.latitude, location.longitude]], 
                               columns=['location', 'latitude', 'longitude'])
        return df
    except GeocoderTimedOut:
        logging.info('TIMED OUT: GeocoderTimedOut: Retrying...')
        print('Error 1')
        sleep(randint(1*100,sleep_sec*100)/100)
        return reverse_geocode(geolocator, loc, sleep_sec)
    except GeocoderServiceError as e:
        logging.info('CONNECTION REFUSED: GeocoderServiceError encountered.')
        logging.error(e)
        return None
    except Exception as e:
        logging.info('ERROR: Terminating due to exception {}'.format(e))
        return None
```
